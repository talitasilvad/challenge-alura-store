### Importação dos dados



```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

df = pd.concat([loja, loja2, loja3, loja4])

print("Tamanho total da tabela: ", df.shape)
df.head(25)
```

# 2. Vendas por Categoria

Neste passo, deve calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

```python
# df.groupby('Categoria do Produto')['Produto'].count
# Passo 1: Agrupamos por Loja E por Categoria (Hierarquia)
# Passo 2: Selecionamos a coluna 'Produto' para contar quantos itens existem
# Passo 3: Executamos o count()

df.groupby('Categoria do Produto')
df.groupby(['Loja', 'Categoria do Produto'])
```

```python
vendas_por_categoria = df.groupby(['Loja', 'Categoria do Produto'])['Produto'].count()
vendas_por_categoria
```

```python
categorias_populares = df.groupby(['Loja', 'Categoria do Produto'])[['Produto']].count().sort_values(by=['Loja','Produto'], ascending=[True, False])
categorias_populares
```

# 3. Média de Avaliação das Lojas


Neste passo, vamos calcular a média das avaliações dos clientes para cada loja. O objetivo **é entender a satisfação dos clientes com os produtos vendidos.**

```python
avaliacao_por_loja = df.groupby('Loja')[['Avaliação da compra']].mean().sort_values('Avaliação da compra', ascending=False)
avaliacao_por_loja
```

# 4. Produtos Mais e Menos Vendidos

Neste passo, deve identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.  Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.

```python
produtos_vendidos = df.groupby(['Loja', 'Produto'])[['Produto']].count()
produtos_vendidos.columns = ['Quantidade_Vendas']

produtos_populares = produtos_vendidos.sort_values(by=['Loja', 'Quantidade_Vendas'], ascending=[True, False])
produtos_populares.groupby('Loja').head(10)
```

```python
produtos_populares.groupby('Loja').tail(10)
```

# 5. Frete Médio por Loja

Neste passo, vamos calcular o custo médio de frete para cada loja. O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas. Sabendo que o cliente que paga o frete.

```python
frete_por_loja = df.groupby('Loja')[['Frete']].mean().sort_values('Frete', ascending=False)
frete_por_loja
```

```python
comparativo = df.groupby('Loja')[['Frete']].agg(['mean', 'median'])
comparativo
```

```python
produtos_outliers = df[df['Frete'] > 200]
tabela_tvs = produtos_outliers.groupby(['Loja', 'Produto'])[['Frete']].count()

tabela_tvs
```

```python
sns.set(style="whitegrid")

faturamento = df.groupby('Loja')['Preço'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Loja',
            y='Preço',
            data=faturamento,
            palette='magma',
            hue='Loja',
            legend=False)
plt.title('Faturamento Total por Loja', fontsize=16)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.xlabel('Lojas', fontsize=12)


plt.show()
```

```python
contagem_categorias = df['Categoria do Produto'].value_counts()

plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(contagem_categorias,
                                   labels=contagem_categorias.index,
                                   autopct='%1.1f%%',
                                   startangle=140,
                                   colors=sns.color_palette('magma'))
for texto in autotexts:
    texto.set_color('white')
    texto.set_weight('bold')
    texto.set_fontsize(12)
plt.title('Distribuição das Vendas por Categoria', fontsize=16)
plt.show()
```

```python
ordem_categorias = df['Categoria do Produto'].value_counts().index

plt.figure(figsize=(12, 6))
sns.countplot(x='Loja',
              hue='Categoria do Produto',
              data=df,
              palette='magma',
              hue_order=ordem_categorias)
plt.title('Vendas de cada Categoria dentro de cada Loja', fontsize=16)
plt.ylabel('Quantidade de Vendas')
plt.legend(title='Categoria', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
df_produtos = produtos_populares.reset_index()
top_produtos = df_produtos.groupby('Loja').head(5)
g = sns.catplot(data=top_produtos,
                x='Quantidade_Vendas',
                y='Produto',
                col='Loja',
                kind='bar',
                palette='magma',
                col_wrap=2,
                sharey=False,
                hue='Produto',
                legend=False)

g.fig.suptitle('Top 5 Produtos Mais Vendidos por Loja', y=1.02, fontsize=16)
g.set_axis_labels("Quantidade Vendida", "Produto")
g.set_titles("{col_name}")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,
                x='Preço',
                y='Frete',
                hue='Loja',
                s=70,
                palette='Set1')

plt.title('Relação: Preço do Produto x Valor do Frete', fontsize=16)
plt.xlabel('Preço do Produto (R$)', fontsize=12)
plt.ylabel('Valor do Frete (R$)', fontsize=12)
plt.axhline(200, color='red', linestyle='--')
plt.text(0, 203, 'Zona dos Outliers (25 TVs)', color='red')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Loja',
            y='Avaliação da compra',
            data=df,
            palette='magma',
            hue='Loja',
            legend=False)

plt.title('Distribuição das Avaliações dos Clientes', fontsize=16)
plt.ylabel('Nota (1 - 5)', fontsize=12)
plt.xlabel('Loja', fontsize=12)

plt.show()
```

## 📊 **Relatório final de análise Estratégica: Alura Store.**

###**1 - Introdução e Objetivo**

O objetivo desta análise foi auxiliar o Sr. João na tomada de decisão estratégica sobre qual unidade da rede Alura Store deve ser vendida para financiar um novo empreendimento.

Para chegar a uma conclusão assertiva, analisamos os dados de vendas, faturamento, frete e satisfação do cliente das quantros lojas, buscando identificar a unidade com menor eficiência operacional e menor potencial de retenção de clientes.


###**2 - Desenvolvimento e Análise de Dados**

Durante a exploração dos dados identificamos padrões cruciais que diferenciam o desempenho das lojas:


💰**Faturamento e Desempenho Financeiro**

* Líder: Loja 1, apresenta o maior faturamento acumulado da rede, aproximadamente **R\$ 1.53 mi**, se consolidando como a principal fonte de receita.

* Última: Loja 4, possui o pior desempenho financeiro, aproximadamente **R\$ 1.38 mi**, com uma diferença significaiva de quase R\$ 150.000,00 a menos que a líder.


📦**Mix de Produtos e Outliers(Logística)**

* Categorias: A categoria Moveis domina o volume de vendas geral das 4 lojas (20.0%), seguida por Eletronicos (18.8%).

* Análise de Frete: Identificamos que a Loja 1 possui o frete médio mais caro, porém nossa análise de disperção revelou que isso é causado por outliers, 9 unidades de 25 TVs Led UHD 4K vendidas com fretes superiores a R\$ 200,00 na Loja 1.

* Eficiência: A loja 4 possui o menor custo de frete médio, o que indica a venda de produtos menos caros, menores ou leves, pórem ela falha em converter isso em volume financeiro suficiente para superar as outras lojas.

⭐**Satisfação do Cliente**

A análise das avaliações revelou uma divisão clara na qualidade do serviço:

* Alta Performance: **Lojas 2 e 3** são as mais bem avaliadas, demonstrando consistência e qualidade no atendimento.
* Baixa Performance: **Lojas 1 e 4** apresentam as piores médias e maior inconsistência nas notas, como evidenciado no gráfico Boxplot.

###**3 - Conclusão e Recomendação**
Com base no cruzamento dos indicadores financeiros e
de satisfação, recomendamos **a venda da Loja 4.**


📉**Justificativa da Decisão**

A Loja 4 é a escolha lógica pois apresenta a pior combinação de indicadores:
1.   Pior Faturamento: É a unidade que menos contribuí para o caixa da empresa.
2.   Baixa Satisfação: Assim como a Loja 1, ela desagarda os clientes, no entanto diferente da Loja 1, ela não compensa esse problema com alto faturamento. Manter uma loja que não gera caixa e ainda detrai a imagem da empresa é insustentável a longo prazo.
